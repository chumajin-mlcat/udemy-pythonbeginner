**お店を作って**材料渡して返してくれるイメージ


例えば **クリーニング屋さん**　をイメージしてください。シャツ渡して、クリーニングして返してくれる

結論から言うと、先ほどの例は以下で定義して一行でできちゃう

```python
def makegraph(csvpath):
  # 復習
  #csvpath = "/content/udemy-pythonbeginner/Session4/sample/00009.csv"
  df = pd.read_csv(csvpath,encoding="shift-jis")
  x = df["温度"]
  y = df["湿度"]
  plt.title("温度と湿度")
  plt.xlabel("温度")
  plt.ylabel("湿度")
  plt.scatter(x,y)
```

**お店を作って**材料渡して返してくれるイメージ


例えば **クリーニング屋さん**　をイメージしてください。シャツ渡して、クリーニングして返してくれる

returnを入れると、出来上がったものを返してくれる(洗濯されたシャツ)

```python
def calc(a):

    b = a + 5 # お店プロセス1 aを材料にbを算出
    c = b ** 2  # bをもとにcを作る

    return c  # cを商品として返す
```

```python
calc(5)
```

```python
calc(100)
```

```python
answer = calc(10000)
answer
```

plt.figure() を関数に入れれば、1枚1枚出力できる

```python
def makegraph(csvpath):
  # 復習
  #csvpath = "/content/udemy-pythonbeginner/Session6/sample/00009.csv"
  df = pd.read_csv(csvpath,encoding="shift-jis")
  x = df["温度"]
  y = df["湿度"]


  plt.figure() # 追加


  plt.title("温度と湿度")
  plt.xlabel("温度")
  plt.ylabel("湿度")
  plt.scatter(x,y)
```

```python
import zipfile
import os

def unzip_file(zip_path, extract_to):
    """ 指定したZIPファイルを指定したフォルダに解凍する """
    os.makedirs(extract_to, exist_ok=True)  # 出力フォルダを作成（存在しない場合のみ）
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)
    print(f"解凍完了: {extract_to}")

# 使用例
zip_path = "/content/udemy-pythonbeginner/Session7/pet.zip"  # 解凍するZIPファイルのパス
extract_to = "/content/udemy-pythonbeginner/Session7"  # 解凍先フォルダ
unzip_file(zip_path, extract_to)
```

場合によっては、windowsの通常の解凍よりもかなり早いです！
