## 4.3 複数のグラフを挿入する

```python
#1 重ねる
df.head(3)
```

```python
x = df["温度"]
y = df["湿度"]
y2 = df["light(ルクス)"]
y3 = df["電圧"]

plt.scatter(x,y,label = "湿度")
plt.scatter(x,y2, label = "light(ルクス)")
plt.scatter(x,y3, label = "電圧")
plt.legend()
```

## 4.4 グラフを分ける

```python
# plt.figure()を挟むとグラフが分けられる.また、plt.figure()でグラフの大きさも変えられる

plt.figure()

plt.scatter(x,y,label = "湿度")

plt.figure()
plt.scatter(x,y2, label = "light(ルクス)")

plt.figure()
plt.scatter(x,y3, label = "電圧")

plt.legend()
```

```python
# subplotで、横に並べたり、縦に並べたり。


plt.subplot(1,3,1)
plt.scatter(x,y,label = "湿度")

plt.subplot(1,3,2)
plt.scatter(x,y2, label = "light(ルクス)")

plt.subplot(1,3,3)
plt.scatter(x,y3, label = "電圧")

plt.legend()
```

```python
# subplotで、横に並べたり、縦に並べたり。


plt.subplot(3,1,1)
plt.scatter(x,y,label = "湿度")

plt.subplot(3,1,2)
plt.scatter(x,y2, label = "light(ルクス)")

plt.subplot(3,1,3)
plt.scatter(x,y3, label = "電圧")

plt.legend()
```
